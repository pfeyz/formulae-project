```python
import pandas as pd
import numpy as np
from dataset import get_manchester
from functionword_tagger import FunctionWordTagger
```

```python
unigrams, bigrams, trigrams = get_manchester("E:/project/Manchester/**/*.xml", 
                                            cached=True, 
                                            gram_sizes=[1, 2, 3])
```

```python
unigrams = unigrams[unigrams.speaker.isin(['CHI', 'MOT'])]
bigrams = bigrams[bigrams.speaker.isin(['CHI', 'MOT'])]
trigrams = trigrams[trigrams.speaker.isin(['CHI', 'MOT'])]
```

```python
bigrams
```

```python
until_bi = bigrams.loc[bigrams['ngram'].str.startswith('until', na=False)]
count_unbi = until_bi.groupby('ngram').size().to_frame().reset_index().rename(columns = {0:'count'})
count_unbi
# Christmas_Day, I, July, June, May, Monday, Saturday, September, Sunday, Wednesday, end, late, later, now, 
# seven, ten, today, tomorrow, twenty, dinner, next, quater, summer, teatime,the(sometimes is) , another
```

```python
until_tri = trigrams.loc[trigrams['ngram'].str.startswith('until', na=False)]
count_untri = until_tri.groupby('ngram').size().to_frame().reset_index().rename(columns = {0:'count'})
count_untri
# until your birthday, until the end, until the morning, until the very last page
```

```python
writer = pd.ExcelWriter('until trigrams.xlsx')
count_untri.to_excel(writer,'Sheet1')
writer.save()
```
