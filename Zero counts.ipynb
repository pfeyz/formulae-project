## Zero counts for each child in each session without filtering

```python
Unigrams
```

```python
d = []
children = unigrams.corpus.unique()
sessions = unigrams.session.unique()
for child in children:
    for sessionss in sessions:
        df = unigrams
        child_ngram = set(df[(df['corpus'] == child) & (df['session'] == sessionss) & (df['speaker'] == 'CHI')]['ngram'].unique())
        adult_ngram = set(df[(df['corpus'] == child) & (df['session'] == sessionss) & (df['speaker'] == 'MOT')]['ngram'].unique())
        in_adult_not_in_child = adult_ngram - child_ngram
        in_child_not_in_adult = child_ngram - adult_ngram
        d.append((child, sessionss , len(in_adult_not_in_child), len(in_child_not_in_adult)))
groups = pd.DataFrame(d, columns = ('corpus', 'session', 'child zero', 'adult zero'))
groups
```
